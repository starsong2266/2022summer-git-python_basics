```python
def lensort(len_dic, i):
  if len(i) <= 40:
    len_dic['<=40'] += 1
  elif 40 < len(i) <= 50:
    len_dic['<=50'] += 1
  elif 50 < len(i) <= 60:
    len_dic['<=60'] += 1
  elif 60 < len(i) <= 70:
    len_dic['<=70'] += 1
  elif 70 < len(i) <= 80:
    len_dic['<=80'] += 1    
  elif len(i) > 80:
    len_dic['>80'] += 1
  return len_dic
```

```python
count = 0
random_noise = '!?+-'
without_noise = 0
len_dic_before = {'<=40':0, '<=50':0, '<=60':0, '<=70':0, '<=80':0, '>80':0}
len_dic_after = {'<=40':0, '<=50':0, '<=60':0, '<=70':0, '<=80':0, '>80':0}
with open(raw_path+data, 'r') as f:
  for i in f:
    len_dic_before = lensort(len_dic_before, i)
with open(raw_path+data, 'r') as f:
  for i in f:
    if (random_noise not in i):
      len_dic_after = lensort(len_dic_after, i)

with open(raw_path+data, 'r') as f:
  for i in f:
    if (random_noise in i):
      for a in i:
        if (a in random_noise):
          i=i.replace(a, '')
      len_dic_after = lensort(len_dic_after, i)
    
      
```
